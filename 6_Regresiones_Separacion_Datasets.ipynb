```python
import pandas as pd
import re
import os
```

# **| Separación del Dataset Grecia**

```python
#Cargar desde un archivo .csv
df = pd.read_csv('Grecia.csv')
Grecia = df.drop(['Unnamed: 0'], axis=1)

Grecia.head()
```

```python
# Separar los datos en diferentes datasets basados en el valor de "room_type"
room_type_groups = Grecia.groupby('room_type')

# Crear un diccionario para almacenar los datasets separados
room_type_datasets = {room_type: data for room_type, data in room_type_groups}

# Mostrar los nombres de los datasets creados
room_type_datasets.keys()
```

```python
for room_type, data in room_type_datasets.items():
    print(f"Dataset: {room_type}")
    display(data.head())
```

```python
# Crear la carpeta del país si no existe
pais = 'Grecia'
carpeta_pais = f"room_type_{pais}"
os.makedirs(carpeta_pais, exist_ok=True)

for room_type, data in room_type_datasets.items():
    # Limpiar el nombre del archivo (eliminar caracteres problemáticos y reemplazar espacios)
    sanitized_name = re.sub(r'[^\w\s]', '', room_type).replace(' ', '_')
    
    # Construir la ruta del archivo dentro de la carpeta del país
    filename = os.path.join(carpeta_pais, f"{sanitized_name}_{pais}.csv")
    
    # Guardar el dataset en un archivo CSV
    data.to_csv(filename, index=False)

    print(f"Archivo guardado en: {filename}")
```
